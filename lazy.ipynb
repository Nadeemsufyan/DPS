```python
from dask.distributed import Client, progress
client = Client(threads_per_worker=4, n_workers=1)
client
```

```python
import time
import random
def inc(x):
    time.sleep(random.random())
    return x + 1
def dec(x):
    time.sleep(random.random())
    return x - 1
def add(x, y):
    time.sleep(random.random())
    return x + y
```

```python
%%time
x = inc(1)
y = dec(2)
z = add(x, y)
z
```

```python
import dask
inc = dask.delayed(inc)
dec = dask.delayed(dec)
add = dask.delayed(add)
```

```python
x = inc(1)
y = dec(2)
z = add(x, y)
z
```

```python
z.visualize(rankdir='LR')
```

```python
z.compute()
```

```python
zs = []
```

```python
%%time
for i in range(256):
    x = inc(i)
    y = dec(x)
    z = add(x, y)
    zs.append(z)
```

```python
zs = dask.persist(*zs)
# trigger computation in the background
```
